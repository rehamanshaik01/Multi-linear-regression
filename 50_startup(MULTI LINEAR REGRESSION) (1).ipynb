```python
#assining target variable
```

```python
y=d1["Profit"]
y
```

```python
x=d1.iloc[:,:4]
x
```

```python
#data preprocessing
```

```python
for columns in x.columns:
    if (columns=="State"):
        from sklearn.preprocessing import LabelEncoder
        le=LabelEncoder()
        x[columns]=le.fit_transform(x[columns]) 
    else:
        from sklearn.preprocessing import StandardScaler
        ss=StandardScaler()
        x[[columns]]=ss.fit_transform(x[[columns]]) 
```

```python
x
```

```python
rmse_test=[]
r2_scores=[]
```

```python
#cross validation with multi linear regression
```

```python
for i in range(1,11):
    from sklearn.model_selection import train_test_split
    x_train,x_test,y_train,y_test=train_test_split(x, y,test_size=0.30,random_state=i)
    from sklearn.linear_model import LinearRegression
    lireg=LinearRegression()
    lireg.fit(x_train,y_train)
    y_pred_test=lireg.predict(x_test)
    from sklearn.metrics import mean_squared_error,r2_score
    rmse_test.append(np.sqrt(mean_squared_error(y_test, y_pred_test)))
    r2_scores.append(r2_score(y_test, y_pred_test))
```

```python
# After fitting the model
import statsmodels.api as sm
residuals = y_test - y_pred_test
fig, ax = plt.subplots(figsize=(6, 4))
sm.qqplot(residuals, fit=True, line="45", ax=ax)
plt.title("Q-Q Plot of Residuals")
plt.show()
```

```python

plt.scatter(y_pred_test, residuals)
plt.axhline(y=0, color="r", linestyle="--")
plt.title("Residuals vs Predicted Values")
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.show()
```

```python
import statsmodels.api as sm

# Add a constant to the predictor variables
x_train_with_const = sm.add_constant(x_train)

# Fit the model
sm_model = sm.OLS(y_train, x_train_with_const).fit()

# Influence plot
influence = sm_model.get_influence()
influence_plot = influence.plot_influence()
plt.title("Influence Plot")
plt.show()
```

```python
print("RMSE of test :",np.mean(rmse_test).round(3))
print("r2_square :",np.mean(r2_scores).round(3))
```
